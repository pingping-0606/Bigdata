````python
# prompt: รวบรวมข้อมูล: ดึงข้อมูลที่เกี่ยวข้อง เช่น Date วันที่ี่(สัปดาห์)
# Temperature อุณหภูมิเฉลี่ยในพื้นที่
# Fuel_Price ราคาน้ำมันในภูมิภาค
# MarkDown ข้อมูลที่ไม่ระบุชื่อที่เกี่ยวข้องกับการลดราคาโปรโมชั่นที่ Walmart กำลังดำเนินการอยู่
# CPI ดัชนีราคาผู้บริโภค
# Type ประเภทร้านค้าไว้แล้ว โดยมี 3 ประเภท  A, B และ C
# Size ขนาดของร้าจัดการกับข้อมูลขาดหาย: ทำการเติมข้อมูลที่หายไปด้วยค่าเฉลี่ย หรือค่าที่เหมาะสมที่สุด

# รวบรวมข้อมูลที่เกี่ยวข้อง
walmart_data = walmart_sales_data[['Date', 'Temperature', 'Fuel_Price', 'CPI', 'Type', 'Size', 'MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5', 'Weekly_Sales']]

# จัดการกับข้อมูลขาดหาย (แทนที่ด้วยค่าเฉลี่ย)
for column in walmart_data.columns:
  if walmart_data[column].dtype == 'float64' or walmart_data[column].dtype == 'int64':
    walmart_data[column].fillna(walmart_data[column].mean(), inplace=True)

# แสดงข้อมูล
walmart_data
````

````python
# prompt: แทนค่า nan เป็น unknow

walmart_sales_data.fillna('Unknown', inplace=True)
walmart_sales_data
````
