    iv. (10 points) สร้างตัวแปร IsAlone ซึ่งมีค่าเป็น 1 ถ้า FamilySize เป็น 0, และมีค่าเป็นศูนย์เมื่อเป็นอื่น

```python
# ตรวจสอบค่าว่างใน DataFrame
titanic.isnull().sum()
```

```python
# ตรวจสอบค่าว่างใน DataFrame
titanic.isnull().sum()
# ลบแถวที่มีค่าว่าง
titanic_cleaned_rows = titanic.dropna()
# ลบคอลัมน์ที่มีค่าว่าง
titanic_cleaned_columns = titanic.dropna(axis=1)
# แทนที่ค่าว่างด้วยค่าคงที่
titanic_filled_constant = titanic.fillna(0)
```

```python
# i.ใช้ one-hot encoding ในการแปลงตัวแปร Embarked
import pandas as pd
# One-hot encoding for Embarked
embarked_encoded = pd.get_dummies(titanic['embarked'], prefix='Embarked')
titanic = pd.concat([titanic, embarked_encoded], axis=1)
embarked_encoded
```

```python
# ii. แปลงตัวแปร Sex โดยให้เพศชายเป็น 0 และเพศหญิงเป็น 1

import pandas as pd
titanic = pd.read_excel('/content/titanic3.xls')
# ลบแถวที่มีค่าว่างในคอลัมน์ sex
titanic.dropna(subset=['sex'], inplace=True)
# แปลงตัวแปร Sex โดยให้เพศชายเป็น 0 และเพศหญิงเป็น 1
titanic['sex'] = titanic['sex'].map({'male': 0, 'female': 1})
titanic['sex']
```

```python
# iii.สร้างตัวแปร IsAlone ซึ่งมีค่าเป็น 1 ถ้า FamilySize เป็น 0, และมีค่าเป็นศูนย์เมื่อเป็นอื่น

import numpy as np
titanic = pd.read_excel('/content/titanic3.xls')
# สร้างตัวแปร FamilySize
titanic['FamilySize'] = titanic['sibsp'] + titanic['parch']

# สร้างตัวแปร IsAlone
titanic['IsAlone'] = np.where(titanic['FamilySize'] == 0, 1, 0)
titanic['IsAlone']
```

```python
# iv.สร้างตัวแปร FamilySize ซึ่งเป็นผลรวมของตัวแปร SibSp และ Parch
import pandas as pd
titanic = pd.read_excel('/content/titanic3.xls')
titanic['FamilySize'] = titanic['sibsp'] + titanic['parch']
titanic['FamilySize']
```

* ทำการ Normalize ข้อมูล \
    i. (10 points) ตัวแปร Fare และ Age ใช้ Min-Max Scaling \
    ii. (10 points) สร้างตัวแปรใหม่ Fare_Per_Person ด้วยการนำตัวแปร Fare มาหารด้วย FamilySize + 1 \
    iii. (10 points) สร้างตัวแปร AgeGroup ที่ระบุกลุ่มอายุของตัวอย่าง โดยแบ่งเป็นกลุ่มดังนี้ [0, 12), [12, 18), [18, 50) และ [50, 100)

```python
import pandas as pd
import numpy as np
pd.cut(np.array([1, 7, 5, 4, 6, 3]), 3)
```

```python
pd.cut(np.array([1, 7, 5, 4, 6, 3]), [0,2,5,8])
```

```python
# i.ตัวแปร Fare และ Age ใช้ Min-Max Scaling
from sklearn.preprocessing import MinMaxScaler

# สร้าง MinMaxScaler object
scaler = MinMaxScaler()

# เลือกคอลัมน์ที่ต้องการทำ Min-Max Scaling
titanic[['fare', 'age']] = scaler.fit_transform(titanic[['fare', 'age']])

# แสดงผล DataFrame หลังจากทำ Min-Max Scaling
print(titanic[['fare', 'age']])
```

```python
# ii.สร้างตัวแปรใหม่ Fare_Per_Person ด้วยการนำตัวแปร Fare มาหารด้วย FamilySize + 1
titanic['FamilySize'] = titanic['sibsp'] + titanic['parch']

titanic['Fare_Per_Person'] = titanic['fare'] / (titanic['FamilySize'] + 1)
titanic['Fare_Per_Person']
```

```python
# iii.สร้างตัวแปร AgeGroup ที่ระบุกลุ่มอายุของตัวอย่าง โดยแบ่งเป็นกลุ่มดังนี้ [0, 12), [12, 18), [18, 50) และ [50, 100)

import pandas as pd
# สร้างตัวแปร AgeGroup
titanic['AgeGroup'] = pd.cut(titanic['age'], bins=[0, 12, 18, 50, 100], labels=['Child', 'Teenager', 'Adult', 'Senior'])
titanic['AgeGroup']
```

* (10 points) นำเสนอการแจกแจงหรือทำความเข้าใจตัวแปรสำคัญในข้อมูล เช่น Age, Fare, and Pclass ด้วยการสร้างฮิสโตแกรมหรือแผนภาพกล่อง

```python
import matplotlib.pyplot as plt
# Histograms for Age, Fare
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(titanic['age'], bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(titanic['fare'], bins=20)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Box plot for Pclass
plt.figure(figsize=(8, 6))
titanic.boxplot(column='pclass', by='survived')
plt.title('Pclass Distribution by Survival')
plt.xlabel('Survived')
plt.ylabel('Pclass')
plt.show()
```

จากกราฟนี้ เราสามารถสรุปได้ว่า

ผู้โดยสารชั้นหนึ่ง (Pclass 1) มีอัตราการรอดชีวิตสูงที่สุด สังเกตได้จากกล่องของกลุ่มที่รอดชีวิต (Survived=1) อยู่ในระดับที่สูงที่สุด แสดงว่าผู้โดยสารส่วนใหญ่ในชั้นหนึ่งสามารถรอดชีวิตได้

ผู้โดยสารชั้นสาม (Pclass 3) มีอัตราการรอดชีวิตต่ำที่สุด สังเกตได้จากกล่องของกลุ่มที่เสียชีวิต (Survived=0) อยู่ในระดับที่สูงที่สุด แสดงว่าผู้โดยสารส่วนใหญ่ในชั้นสามเสียชีวิต

ไม่มี Outlier ที่ชัดเจน ไม่พบจุดที่อยู่ห่างจากกล่องมากเกินไป ซึ่งอาจเป็นเพราะข้อมูลเกี่ยวกับชั้นโดยสารมีการแบ่งกลุ่มที่ชัดเจนอยู่แล้ว

* เมื่อเตรียมข้อมูลพร้อมแล้ว ให้หาความสัมพันธ์ของตัวแปรต่อไปนี้จากข้อมูล ซึ่งอาจนำเสนอด้วยสถิติและแผนภูมิหรือกราฟก็ได้: \
    i. (10 points) survived และ pclass \
    ii. (10 points) survived และ sex \
    iii. (10 points) survived และ age \
    iv. (10 points) จำนวนสมาชิกครอบครัวที่รอดชีวิต และ Fare

```python
#i.ความสัมพันธ์ของ survived และ pclass
import pandas as pd
import matplotlib.pyplot as plt
# Calculate the survival rate for each Pclass
survival_rate_by_pclass = titanic.groupby('pclass')['survived'].mean()
print(survival_rate_by_pclass)

# สร้างแผนภูมิแท่งเพื่อให้เห็นภาพความสัมพันธ์
plt.figure(figsize=(8, 6))
survival_rate_by_pclass.plot(kind='bar')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class (Pclass)')
plt.ylabel('Survival Rate')
plt.show()

# สร้างตาราง
contingency_table = pd.crosstab(titanic['survived'], titanic['pclass'], normalize='index')
print(contingency_table)

# สร้าง heatmap เพื่อแสดงภาพความสัมพันธ์
plt.figure(figsize=(8, 6))
plt.imshow(contingency_table, cmap='viridis', interpolation='nearest')
plt.colorbar()
plt.xticks([0, 1, 2], ['Pclass 1', 'Pclass 2', 'Pclass 3'])
plt.yticks([0, 1], ['Died', 'Survived'])
plt.title('Survival Rate by Passenger Class (Heatmap)')
plt.xlabel('Passenger Class')
plt.ylabel('Survived')
plt.show()
```

จากกราฟนี้ เราสามารถสรุปได้ว่าชั้นโดยสารเป็นปัจจัยสำคัญที่ส่งผลต่อโอกาสในการรอดชีวิตของผู้โดยสารบนเรือไททานิก ผู้โดยสารในชั้นสูงมีโอกาสรอดชีวิตมากกว่าผู้โดยสารในชั้นต่ำ ซึ่งสะท้อนให้เห็นถึงความไม่เท่าเทียมกันในสังคมในยุคนั้น

```python
#ii.ความสัมพันธ์ของ survived และ sex
import pandas as pd
import matplotlib.pyplot as plt
# คำนวณอัตราการรอดชีวิตของแต่ละเพศ
survival_rate_by_sex = titanic.groupby('sex')['survived'].mean()
print(survival_rate_by_sex)

# สร้างแผนภูมิแท่งเพื่อให้เห็นภาพความสัมพันธ์
plt.figure(figsize=(8, 6))
survival_rate_by_sex.plot(kind='bar')
plt.title('Survival Rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.show()

# สร้างตาราง
contingency_table = pd.crosstab(titanic['survived'], titanic['sex'], normalize='index')
print(contingency_table)

# สร้าง heatmap เพื่อแสดงภาพความสัมพันธ์
plt.figure(figsize=(8, 6))
plt.imshow(contingency_table, cmap='viridis', interpolation='nearest')
plt.colorbar()
plt.xticks([0, 1], ['Male', 'Female'])
plt.yticks([0, 1], ['Died', 'Survived'])
plt.title('Survival Rate by Sex (Heatmap)')
plt.xlabel('Sex')
plt.ylabel('Survived')
plt.show()

```

สรุปผลจากกราฟเห็นว่า
เพศหญิงมีอัตราการรอดชีวิตสูงกว่าเพศชายอย่างเห็นได้ชัด จากกราฟจะเห็นได้ว่าคอลัมน์ที่แสดงถึงเพศหญิงมีความสูงมากกว่าคอลัมน์ที่แสดงถึงเพศชายอย่างมาก หมายความว่าผู้หญิงมีโอกาสรอดชีวิตจากเหตุการณ์เรือไททานิกจมมากกว่าผู้ชาย

```python
#iii.ความสัมพันธ์ของ survived และ age

import matplotlib.pyplot as plt
# สร้าง scatter plot เพื่อแสดงความสัมพันธ์ระหว่าง survived และ age
plt.figure(figsize=(8, 6))
plt.scatter(titanic['age'], titanic['survived'], alpha=0.5)
plt.title('Relationship between Survived and Age')
plt.xlabel('Age')
plt.ylabel('Survived')
plt.show()

# สร้าง box plot เพื่อแสดงการกระจายตัวของอายุตามสถานะการรอดชีวิต
plt.figure(figsize=(8, 6))
titanic.boxplot(column='age', by='survived')
plt.title('Age Distribution by Survival')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```

สรุปผลจากกราฟเห็นว่า
อายุเฉลี่ยของผู้รอดชีวิตและผู้เสียชีวิตมีความใกล้เคียงกัน จากกราฟจะเห็นได้ว่ากล่องของทั้งสองกลุ่ม (รอดชีวิตและเสียชีวิต) อยู่ในตำแหน่งที่ใกล้เคียงกัน ซึ่งหมายความว่าอายุเฉลี่ยของผู้รอดชีวิตและผู้เสียชีวิตไม่ได้แตกต่างกันมากนัก

มีผู้สูงอายุเป็น Outlier ในทั้งสองกลุ่ม มีจุดที่แสดงถึงอายุที่สูงมากอยู่เหนือกล่องทั้งในกลุ่มที่รอดชีวิตและเสียชีวิต ซึ่งอาจบ่งบอกว่ามีผู้สูงอายุปนอยู่ทั้งในกลุ่มผู้รอดชีวิตและผู้เสียชีวิต

```python
#iv.ความสัมพันธ์ของ จำนวนสมาชิกครอบครัวที่รอดชีวิต และ Fare

import matplotlib.pyplot as plt
# สร้าง scatter plot เพื่อแสดงความสัมพันธ์ระหว่าง FamilySize และ Fare
plt.figure(figsize=(8, 6))
plt.scatter(titanic['FamilySize'], titanic['fare'], alpha=0.5)
plt.title('Relationship between FamilySize and Fare')
plt.xlabel('FamilySize')
plt.ylabel('Fare')
plt.show()

# สร้าง box plot เพื่อแสดงการกระจายตัวของค่าโดยสารตามจำนวนสมาชิกครอบครัว
plt.figure(figsize=(8, 6))
titanic.boxplot(column='fare', by='FamilySize')
plt.title('Fare Distribution by FamilySize')
plt.xlabel('FamilySize')
plt.ylabel('Fare')
plt.show()
```

สรุปผลจากกราฟ
- ค่าโดยสารเฉลี่ยของครอบครัวขนาดเล็กต่ำกว่าครอบครัวขนาดใหญ่: จากกราฟจะเห็นได้ว่ากล่องของกลุ่มที่มีขนาดครอบครัวน้อย (เช่น 0, 1, 2) อยู่ในตำแหน่งที่ต่ำกว่ากลุ่มที่มีขนาดครอบครัวใหญ่ (เช่น 4, 5, 6) ซึ่งหมายความว่าโดยเฉลี่ยแล้ว ครอบครัวขนาดเล็กจะจ่ายค่าโดยสารน้อยกว่าครอบครัวขนาดใหญ่

- มีผู้โดยสารบางรายในครอบครัวขนาดใหญ่จ่ายค่าโดยสารสูงมาก: มีจุดที่แสดงถึงค่าโดยสารที่สูงมากอยู่เหนือกล่องในกลุ่มที่มีขนาดครอบครัวใหญ่ ซึ่งอาจบ่งบอกว่ามีครอบครัวขนาดใหญ่บางครอบครัวที่ซื้อตั๋วโดยสารในชั้นสูง

- ครอบครัวขนาดเล็กมีการกระจายตัวของค่าโดยสารค่อนข้างกว้าง: จากขนาดของกล่องและจำนวน Outlier จะเห็นได้ว่าครอบครัวขนาดเล็กมีการกระจายตัวของค่าโดยสารค่อนข้างกว้าง หมายความว่าค่าโดยสารของครอบครัวขนาดเล็กมีความแตกต่างกันมาก
